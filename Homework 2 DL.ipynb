Problem 4: Linear equations have the general format of 
$$y=mx + b$$
where
$$m\,=\,slope$$
$$b\,=\,y-intercept$$

In the first cell, write a function that returns the y value when a user inputs the slope, $y$-intercept, and $x$ value for a linear equation.

In the second cell, create a list that contains the $x$ values, a list of corresponding $y$ values using your function, and create a line plot using the lists. The slope of the line for this data is 2.5 and the y-intercept is 0.15.

$x$ = 0,5,10,15,20,25,30

Include commenting in your code.

```python
def calc_y(x,m,intercept):
    return m*x + intercept   ## linear equation
```

```python
m = 2.5
intercept = 0.15
x = [0,5,10,15,20,25,30]
y = [0,0,0,0,0,0,0]  # create dummy list for y values

# populate y list with function output
y[0] = calc_y(x[0],m,intercept)
y[1] = calc_y(x[1],m,intercept)
y[2] = calc_y(x[2],m,intercept)
y[3] = calc_y(x[3],m,intercept)
y[4] = calc_y(x[4],m,intercept)
y[5] = calc_y(x[5],m,intercept)
y[6] = calc_y(x[6],m,intercept)

# Create plot
plt.plot(x,y)
plt.title('Linear equation plot')
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(ymin=0)  # set y origin at 0
plt.text(6,5,'y=2.5x + 0.15')
plt.show()
```
