Problem 1: In lab you have a container full of $0.5\ \mathrm{moles}$ of nitrogen gas. You can change the volume of the container while holding the temperature constant at room temperature (about $293\ \mathrm{K}$). You start the volume at $10.0\ \mathrm{L}$ and decrease it by $1.0\ \mathrm{L}$ until you get down to $1.0\ \mathrm{L}$, measuring the (fictitious) pressure at each volume:

|Volume (L)|Pressure (atm)|
| :--: | :--: |
|10.0|1.203|
|9.0|1.333|
|8.0|1.504|
|7.0|1.720|
|6.0|1.998|
|5.0|2.407|
|4.0|3.010|
|3.0|3.999|
|2.0|6.011|
|1.0|12.021|

Make a plot of pressure ($y$-axis) vs. volume ($x$-axis). For your plot, have a dotted, 3 point-thick black line with a square marker (use keyword input parameters).  Start both axes at the minimum values and end them on their maximum values.  Place $x$ tick marks on the even values of volume.  Label your axes and title the plot appropriately (make sure to include that the gas is nitrogen).  Put the number of moles on the plot using the `text` function.

```python
import numpy as np
import matplotlib.pyplot as plt

volume = np.arange(10.0,0.0,-1.0)
pressure = np.array([1.203,1.333,1.504,1.720,1.998,2.407,3.010,3.999,6.011,12.021])

plt.plot(volume,pressure,':ks',linewidth=3)
plt.xlim(0,10.0)
plt.ylim(1.203,12.021)
plt.xticks([10.0,8.0,6.0,4.0,2.0,0.0])
plt.xlabel('Volume (L)')
plt.ylabel('Pressure (atm)')
plt.title('Pressure vs Volume of Nitrogen Gas')
plt.text(5,8,'0.5 moles of nitrogen gas')
plt.show()
```

Problem 2:  Problem for Physics 5-3: You have (fictitious) data on wave vertical displacement ($y$) versus wave number ($k$) times position ($x$), for two water waves at time t = 0:

|$kx$ (rad)|Wave 1, $y$ (m)|Wave 2, $y$ (m)|
| :--: | --: | --: |
|0.0|0.000|0.212|
|$\frac{\pi}{4}$|0.212|0.300|
|$\frac{\pi}{2}$|0.300|0.212|
|$\frac{3\pi}{4}$|0.212|0.000|
|$\pi$|0.000|-0.212|
|$\frac{5\pi}{4}$|-0.212|-0.300|
|$\frac{3\pi}{2}$|-0.300|-0.212|
|$\frac{7\pi}{4}$|-0.212|0.000|
|$2\pi$|0.000|0.212|


First, calculate the superposition of the two waves (the sum of the vertical displacements for each $kx$). Second, make a single plot of the curves of both waves and the superposition with $kx$ on the $x$-axis and vertical displacement on the $y$-axis. Give the plot the title "Superposition of Two Water Waves with Different Phases", label the axes appropriately, and use the labels "Wave 1," "Wave 2," and "Superposition of Waves" for the legend. Use a square marker with a dashed-dot blue line for wave 1, a circle marker with a dashed red line for wave 2, and a diamond with a solid green line for the superposition.

```python
kx = np.array([0.0,(np.pi/4),(np.pi/2),(3*np.pi/4),(np.pi),(5*np.pi/4),(3*np.pi/2),(7*np.pi/4),(2*np.pi)])
wave1 = np.array([0.000,0.212,0.300,0.212,0.000,-0.212,-0.300,-0.212,0.000])
wave2 = np.array([0.212,0.300,0.212,0.000,-0.212,-0.300,-0.212,0.000,0.212])
superposition = wave1 + wave2

plt.plot(kx,wave1,'-.bs',label='Wave 1')
plt.plot(kx,wave2,'--ro',label='Wave 2')
plt.plot(kx,superposition,'-gD',label='Superposition of Waves')
plt.xlabel('kx (rad)')
plt.ylabel('Vertical Displacement (m)')
plt.title('Superposition of Two Water Waves with Different Phases')
plt.legend()
plt.show()
```

A fictitious set of data is provided in the table.

|Dye Concentration (mM)|Absorbance|
|:--:|:--:|
|0.020563|0.1248|
|0.061688|0.2368|
|0.185063|1.160|
|0.370127|2.258|

Linear Regression Result:  y = 6.2971x - 0.0586

Create a plot that contains the following:

Problem 5: A Lineweaver-Burk plot can be used to identify the type of inhibition an enzyme might be under. This type of plot can be created using following equation:

$$\frac{1}{V} = \frac{1}{[S]} * \frac{K_{m}}{V_{max}} + \frac{1}{V_{max}}$$

$V$ = reaction velocity  
$[S]$ = substrate concentration  
$K_{m}$ = Michaelis-Menten constant  
$V_{max}$ = maximal velocity  

You have the kinetic parameters of an enzyme under controlled conditions as well as inhibition. The enzyme has a $K_{m}$ of 2.95 mM, a $V_{max}$ of 39.6 mmol/min, and a $V_{max}'$ of 21.9 mmol/min under inhibition. Using these parameters, the equation above, and array syntax, calculate $\frac{1}{V}$ when $\frac{1}{[S]}$ equals -0.34, -0.17, 0, 0.17, 0.34, 0.51, 0.68, and 0.85 mM$^{-1}$ for both $V_{max}$ and $V_{max}'$. Plot $\frac{1}{[S]}$ on the x-axis and $\frac{1}{V}$ on the y-axis. Because you cannot have a negative substrate concentration, the line calculated using $V_{max}$ should be a solid black line for $\frac{1}{[S]} \ge 0$ and a dotted black line for $\frac{1}{[S]} < 0$, while the line calculated using $V_{max}'$ should be a solid red line for $\frac{1}{[S]} \ge 0$ and a dotted red line for $\frac{1}{[S]} < 0$. Use array splicing to accomplish the split in line style. Make sure the red and black lines do not have unnecessarily large gaps. Label the axes and title the graph "Noncompetitive inhibition". Save your figure to a file with 150 dpi. 

```python
km = 2.95
vmax = 39.6
vimax = 21.9
inverse_s = np.array([-0.34,-0.17,0,0.17,0.34,0.51,0.68,0.85])

inverse_v = inverse_s * (km/vmax) + (1/vmax)
print(inverse_v)

inverse_vi = inverse_s * (km/vimax) + (1/vimax)
print(inverse_vi)

plt.plot(inverse_s[2:],inverse_v[2:],'-k')
plt.plot(inverse_s[0:3],inverse_v[0:3],':k')
plt.plot(inverse_s[2:],inverse_vi[2:],'-r')
plt.plot(inverse_s[0:3],inverse_vi[0:3],':r')
plt.title('Noncompetitive Inhibition')
plt.xlabel('Inverse reaction velocity')
plt.ylabel('Inverse substrate concentration (mM^-1)')
plt.show()
plt.savefig('file.png',dpi=150)
```

```python
sepal_width = [3.8, 4.0, 4.1, 3.5, 3.5, 3.4,
               2.9, 3.0, 2.5, 2.9, 3.0, 2.9,
               3.2, 3.2, 3.2, 3.0, 2.8, 3.1]
petal_width = [0.3, 0.2, 0.1, 0.2, 0.3, 0.3,
               1.3, 1.4, 1.1, 1.4, 1.5, 1.3,
               2.3, 1.8, 2.3, 1.8, 2.2, 2.1]
```
